```python
import fitz  # PyMuPDF
import google.generativeai as genai
from google.colab import files
import os
```

```python
uploaded = files.upload()
pdf_path = list(uploaded.keys())[0]
```

```python
def extract_text_from_pdf(path):
    doc = fitz.open(path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

document_text = extract_text_from_pdf(pdf_path)
```

```python
GEMINI_API_KEY = "AIzaSyA_V1oUfkmm4KGR6qSP5zD7SGhRBFARr40"  # Replace with your actual key
genai.configure(api_key=GEMINI_API_KEY)
model = genai.GenerativeModel("gemini-2.0-flash")
```

```python
framework_prompt = f"""
Read this content about Prompt Engineering and extract it into a structured framework of key concepts.
Respond with bullet points grouped by:
- Definitions
- Techniques
- Applications
- Risks
- Examples

Content:
{document_text[:8000]}  # You may truncate if needed to avoid token limits
"""

framework_response = model.generate_content(framework_prompt)
framework = framework_response.text
print("📌 Framework Extracted:\n")
print(framework)
```

```python
mcq_prompt = f"""
Based on the following content about Prompt Engineering, generate 10 multiple-choice questions.
Each question should have:
- 1 correct answer
- 3 plausible distractors
- Label the correct answer clearly.

Content:
{framework}
"""

mcq_response = model.generate_content(mcq_prompt)
print("\n📘 Multiple-Choice Questions:\n")
print(mcq_response.text)
```
